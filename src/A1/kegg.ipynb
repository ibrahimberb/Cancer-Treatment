```python
def load_pathways():
    with open("kegg_data.txt", 'r')as ifile:
        pathways={}
        for line in ifile:
            line = line.rstrip('\n')
            p=line.split('\t\t\t')[0] #pathway ID
            genes=set(line.split('\t\t\t')[1].split(",") [:-1])
            for g in genes:
                if g not in pathways:
                    pathways[g]=set()
                    pathways[g].add(p)
                else:
                    pathways[g].add(p)
    return pathways


pathways = load_pathways()
```

```python
numbers = {}
for key, value in pathways.items():
    # print(f"{key} : {len(value)}")
    numbers[key] = len(value)
```

```python
dict(sorted(numbers.items(), key=lambda item: item[1], reverse=True))
```
